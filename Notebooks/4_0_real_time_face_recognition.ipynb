# Creating a real-time face recognition program

We are going to put together the code that we have writeen so far to create a real-time face recognition program. This program will use the webcam in our computer for facial recognition, and to authenticate whether the person sitting in front of the webcam is indeed you.

To do so, the program needs to do he following:

1. Train a Siamese neural network for facial recognition
2. Use the webcam to capture a true image of the authorized user. This is the onboarding process of the facial recognition system.
3. Subsequently, when a user wants to unlock the program, use the pre-trained Siamse neural network from the step 1 and the true image from step 2 to authenticate the user.
